```python
grouped_matches = matches.groupby("team")
```

```python
group = grouped_matches.get_group("Manchester City").sort_values("date")
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]

rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling
```

```python
def make_predictions(data, predictors):
    train = data[data["date"] < '2022-01-01']
    test = data[data["date"] > '2022-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    error = precision_score(test["target"], preds)
    return combined, error
```

```python
combined, error = make_predictions(matches_rolling, predictors + new_cols)
```

```python
class MissingDict(dict):
    __missing__ = lambda self, key: key

map_values = {"Brighton and Hove Albion": "Brighton", "Manchester United": "Manchester Utd", "Newcastle United": "Newcastle Utd", "Tottenham Hotspur": "Tottenham", "West Ham United": "West Ham", "Wolverhampton Wanderers": "Wolves"} 
mapping = MissingDict(**map_values)
```

```python
combined["new_team"] = combined["team"].map(mapping)
```

```python
merged = combined.merge(combined, left_on=["date", "new_team"], right_on=["date", "opponent"])
merged
```

```python
# --- ✅ User-friendly output section ---

# Calculate overall model accuracy
overall_accuracy = (merged["actual_x"] == merged["predicted_x"]).mean() * 100
print(f"\n🎯 Overall Model Accuracy: {overall_accuracy:.2f}%\n")

# Iterate over the merged matches to print readable summaries
for i, row in merged.iterrows():
    date = row["date"].strftime("%Y-%m-%d")
    team = row["team_x"]
    opponent = row["opponent_x"]
    
    # Decode actual and predicted outcomes
    actual_result = "Won" if row["actual_x"] == 1 else "Lost" if row["actual_x"] == 0 else "Draw"
    predicted_result = "Won" if row["predicted_x"] == 1 else "Lost" if row["predicted_x"] == 0 else "Draw"
    
    # Compare prediction accuracy
    status = "✅ Correct" if row["actual_x"] == row["predicted_x"] else "❌ Incorrect"
    
    print(f"🏟️  Match: {team} vs {opponent}  ({date})")
    print(f"   ➤ Actual Result: {team} {actual_result}")
    print(f"   ➤ Model Prediction: {team} {predicted_result}")
    print(f"   ➤ Prediction Status: {status}")
    print("-" * 65)

# --- Optional: Summary of most incorrect predictions ---
incorrect = merged[merged["actual_x"] != merged["predicted_x"]]
print(f"\n⚠️  Total Incorrect Predictions: {len(incorrect)}\n")

# Show top 5 incorrect ones
print("📋 Sample of Incorrect Predictions:")
print(incorrect[["date", "team_x", "opponent_x", "actual_x", "predicted_x"]].head())
```

```python
# --- ✅ Total Correct Predictions ---
correct = merged[merged["actual_x"] == merged["predicted_x"]]
print(f"\n✅ Total Correct Predictions: {len(correct)}\n")

# Show top 5 correct ones
print("📋 Sample of Correct Predictions:")
print(correct[["date", "team_x", "opponent_x", "actual_x", "predicted_x"]].head())
```
